# Calclulating moving averages data in DF

```python
df3=pd.read_csv('/Users/anoop.sharma/Downloads/OPTIDX_BANKNIFTY_CE_17-04-2020_TO_23-04-2020.csv',index_col=1)
df3=df3[df3['Open Int']!=0]
df3
```

```python

df3['Open Int'].plot(color='green',figsize=(16,10))
df3.rolling(window=20).mean()['Open Int'].plot()
```

```python
df3['LTP'].plot(figsize=(16,10))
df3.rolling(window=20).mean()['LTP'].plot()
```

```python
# With legend as optin

df3['Open Int 5 day MA']=df3['Open Int'].rolling(window=5).mean()
df3[['Open Int 5 day MA','Open Int']].plot(figsize=(16,10))
```
